## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Plot

```python
products = ['Mouse' ,'Keyboard' , 'Computer' ,'Cpu']
price = [115,135,120,146]

plt.plot(products,price , color = 'r' , linestyle = '--' , linewidth = '3') 
# supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.xlabel('Products')
plt.ylabel('Prices')
plt.title('Products Line Plot')
plt.show()
```

```python
products = ['Mouse' ,'Keyboard' , 'Computer' ,'Cpu']
price = [115,135,120,146]

plt.plot(products,price , marker = 'o' , markersize = '10' , markerfacecolor = 'r') 
plt.xlabel('Products')
plt.ylabel('Prices')
plt.title('Products Line Plot')
plt.grid(color = 'y' , linestyle = ':')
plt.show()
```

#### Multiple Line Plot

```python
products = ['Mouse' ,'Keyboard' , 'Computer' ,'Cpu']
price = [115,135,120,146]
quantity = [25,10,20,19]

plt.plot(products,price , label = 'Price') 
plt.plot(products,quantity , label = 'Quantity' , linestyle = '--') 
plt.legend()
plt.grid()
plt.show()
```

#### Subplot

```python
products = ['Mouse' ,'Keyboard' , 'Computer' ,'Cpu']
price = [115,135,120,146]
quantity = [25,10,20,19]

plt.subplot(2,1,1) # (rows , column , index) - First subplot
plt.plot(products,price , label = 'Price') 
plt.grid()
plt.title('Prices')

plt.subplot(2,1,2) # (rows , column , index) - Second subplot
plt.plot(products,quantity , label = 'Quantity' ) 
plt.title('Quantity')
plt.grid()

plt.tight_layout()
plt.show()
```

#### saving Figure

```python
products = ['Mouse' ,'Keyboard' , 'Computer' ,'Cpu']
price = [115,135,120,146]
quantity = [25,10,20,19]

plt.subplot(2,1,1) # (rows , column , index) - First subplot
plt.plot(products,price , label = 'Price') 
plt.grid()
plt.title('Prices')

plt.subplot(2,1,2) # (rows , column , index) - Second subplot
plt.plot(products,quantity , label = 'Quantity' ) 
plt.title('Quantity')
plt.grid()

plt.savefig('subplot.png')
plt.tight_layout()
plt.show()
```

## Bar Plot

```python
products = ['Mouse' ,'Keyboard' , 'Computer' ,'Cpu']
price = [115,135,120,146]
quantity = [25,10,20,19]

plt.bar(products,price , color=['r' , 'g','y'] , width = 0.3)

bars = plt.bar(products,price , color=['r' , 'g','y'] , width = 0.3)

plt.xlabel('Products')
plt.ylabel('Prices')
plt.title('Products Bar Plot')
plt.grid(axis='x' , linestyle = ':')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2 , yval , round(yval,2) , ha = 'center' ,va ='bottom' )
    
plt.show()
```

```python
plt.barh(products,price , color=['r' , 'g','y'] )
plt.show()
```

#### Multiple Bar Plot

```python
products = ['Mouse' ,'Keyboard' , 'Computer' ,'Cpu']
price = [115,135,120,146]
quantity = [25,10,20,19]
width = 0.3

product_index = np.arange(len(products))
product_index_set = [j+width for j in product_index]

plt.bar(product_index,price , color='r' , width = width , label = 'Price')
plt.bar(product_index_set,quantity , color='b' , width = width , label = 'Quantity')
plt.xticks(product_index + width/2, products , rotation = 20)
plt.legend()
plt.show()
```

#### Subplot 

```python
products = ['Mouse' ,'Keyboard' , 'Computer' ,'Cpu']
price = [115,135,120,146]
quantity = [25,10,20,19]

fig , axs = plt.subplots(1 ,2 , figsize = (11,4))

# First subplot
axs[0].bar(products,price , color = 'lightblue')
axs[0].set_title('Price Plot')
axs[0].set_xlabel('Products')
axs[0].set_ylabel('Price')

# Second subplot
axs[1].bar(products,quantity ,  color = 'lightgreen')
axs[1].set_title('Quantity Plot')
axs[1].set_xlabel('Products')
axs[1].set_ylabel('Quantity')

plt.tight_layout()
plt.show()
```
