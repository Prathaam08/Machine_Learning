## Advance Operation with Business example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [ restaurant_id , 2021 ,2022 , 2023 , 2024 ]
sales_data = np.array([
    [1 , 150000 , 180000 , 220000 , 250000,],   # resaurant 1
    [2 , 120000 , 140000 , 160000 , 190000,],   # resaurant 2
    [3 , 200000 , 230000 , 260000 , 300000,],   # resaurant 3
    [4 , 180000 , 210000 , 240000 , 270000,],   # resaurant 4
    [5 , 160000 , 185000 , 200000 , 230000,],   # resaurant 5
])

print("======= Zomato sales analysis =======")
print("sales data shape : " , sales_data.shape)
print("Sales data for 1st three restaurant :\n", sales_data[:3])
```

```python
# total sales per year
print(np.sum(sales_data, axis=0 ))
yearly_sales = np.sum(sales_data[: , 1 :] ,axis=0)
print("Yearly sales : " ,yearly_sales)

# Minimum sales per restaurant
min_sales = np.min(sales_data[: , 1: ] ,axis=1)
print()
print(sales_data)
print("\nMinimum sales per restaurant : ", min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[: , 1 :] ,axis=0)
print("\nMaximum sales per year : " ,max_sales)

# Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:] ,axis=1)
print("\nAverage sales per restaurant : " ,avg_sales)

# cumalative sales per restaurant
cum_sales = np.cumsum(sales_data[: ,1 :] ,axis=1)
print("\nCumalative sales per restaurant :\n" , cum_sales)

plt.figure(figsize=(8,6))
plt.plot(np.mean(cum_sales , axis=0))
plt.title("Average cumaltive sales across all restuarant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
